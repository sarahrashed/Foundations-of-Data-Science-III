```python
# import the important packages
import pandas as pd  # library used for data manipulation and analysis
import numpy as np  # library used for working with arrays
import matplotlib.pyplot as plt  # library for plots and visualizations
import seaborn as sns  # library for visualizations

%matplotlib inline

import scipy.stats as stats  # this library contains a large number of probability distributions as well as a growing library of statistical functions
```


Central Limit Theorem states that if we independently draw multiple samples from a population, take the mean of each sample and plot these (sample means), then the plot will tend to normal distribution as the size of samples increases, **regardless** of the shape of the population distribution.

### Let's watch CLT in action using a python simulation

#### Here is a uniform distribution (which is most definitely *not normal*)

```python
# importing the required function
from scipy.stats import uniform

# setting the seed for reproducibility
np.random.seed(1)
uniform_pop = uniform.rvs(scale=10,size=100000) #TODO: create a uniform distribution population on [0,10] of size 100000
# visualizing the uniform distribution
plt.hist(uniform_pop)
plt.title("Uniform Distribution Population")
plt.xlabel("X")
plt.ylabel("Count")
plt.show()
```

#### Define a function that creates the sampling distribution of the sample means and a function that plots it

```python
# define a function that creates the sampling distribution from a certain population with sample size = n and 
# number of samples = 500
def sampling_distrib(n, population):
    # set the seed for reproducibility
    np.random.seed(1)
    sample_means = [] #TODO: create an empty list to store sample means
    # iterate the loop to draw multiple samples
    for j in range(500):
        sample = np.random.choice(population, size=n) #TODO: draw a sample of size n from pop
        sample_mean = sum(sample)/n #TODO: calculate the sample mean
        # append the sample mean to the sample_means list
        sample_means.append(sample_mean)
    return (sample_means)
```

```python
# define a function that plots the histogram of sample means
def plot_sample_means(n, population):
    sample_means = sampling_distrib(n,population) #TODO: find the sample means using the above sampling_distrib function
    sns.displot(sample_means, kde=True)
    plt.title("Distribution of Sample Means for n = " + str(n))
    plt.xlabel("Sample mean")
    plt.ylabel("Count")
    plt.show()
```

#### Let's plot a sampling distribution from uniform_pop population (sample size = 5, number of samples = 500)

*   draw a sample of size 5, so n=5
*   get the mean of these 5 observations, i.e - sample mean
*   repeat the above 2 steps 500 times, so that we get 500 sample means (done in our sampling_distrib function)

Let's observe the shape of this sampling distribution.

```python
# plot a sampling distribution from uniform_pop with sample size = 5 
plot_sample_means(5, uniform_pop)
```

#### Increase the sample size to 15 (n = 15)

```python
# plot a sampling distribution from uniform_pop with sample size = 15 
plot_sample_means(15, uniform_pop)
```

#### Increase the sample size to 30 (n = 30)

```python
# plot a sampling distribution from uniform_pop with sample size = 30
plot_sample_means(30, uniform_pop)
```

#### Increase the sample size to 50 (n = 50)

```python
# plot a sampling distribution from uniform_pop with sample size = 50
plot_sample_means(50, uniform_pop)
```

#### Insight

* Observe how the sampling distribution moves closer to normality as the sample size increases.

#### When the population distribution is normal

```python
# import the required function
from scipy.stats import norm

# set the seed for reproducibility
np.random.seed(1)
normal_pop = norm.rvs(loc=0,scale=1,size=100000) #TODO: create a normal distribution population with mean 0 and standard deviation 1 of size 100000
# visualize the normal distribution
plt.hist(normal_pop, 200)
plt.title("Normal Distribution Population")
plt.xlabel("X")
plt.ylabel("Count")
plt.show()
```

#### Let's create a sampling distribution from this population (sample size = 5, number of samples = 500)

*   draw a sample of size 5, so n=5
*   get the mean of these 5 observations, i.e - sample mean
*   repeat the above 2 steps 500 times, so that we get 500 sample means (done in our sampling_distrib function) 

Let's observe the shape of this sampling distribution.

```python
# plot a sampling distribution from uniform_pop with sample size = 5 
plot_sample_means(5, normal_pop)
```

#### Insight

* When the population distribution is normal, sampling distribution is close to normal for even the smaller sampling sizes like n = 5.

#### Let's see if CLT works if the population distribution is exponential (which again is clearly *not normal*)

```python
# import the required function
from scipy.stats import expon

# set the seed for reproducibility
np.random.seed(1)
exp_pop = expon.rvs(size=100000) #TODO: create a exponential distribution population in the standardized form of size 100000
# visualize the exponential distribution
plt.hist(exp_pop, 200)
plt.title("Exponential Distribution Population")
plt.xlabel("X")
plt.ylabel("Count")
plt.show()
```
